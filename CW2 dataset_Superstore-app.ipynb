> Removing Outliers

There is no one-size-fits-all solution for dealing with outliers in a dataset. Depending on the context and purpose of the analysis, it may be appropriate to remove or keep some outliers (Sharda, 2021).As previously stated reasons for outlier removal, the following methods can be used to remove those outliers:

z-score method: This method sets an outlier threshold by identifying any data point with a z-score greater than 3. (i.e., 3 standard deviations away from the mean). Using boolean indexing, these outliers are then removed from the dataset. Finally, the summary statistics for the modified'Store' dataset's 'Quantity' column are printed.

Interquartile range (IQR) method: This method involves calculating the IQR of the data and removing data points that fall outside a certain multiple of the IQR. For example, we can remove all data points that fall outside 1.5 times the IQR. Here is an example code to remove outliers using the IQR method

```
# Calculate z-score
z = np.abs((Store['Quantity'] - Store['Quantity'].mean()) / Store['Quantity'].std())

# Identify outliers
outliers = z > 3

# Remove outliers
Store = Store[~outliers]

# Print summary statistics
print(Store['Quantity'].describe())
```

```
# Calculate the quartiles of the Sales column
# q1, q3 = np.percentile(Store['Sales_r'], [25, 75])
# iqr = q3 - q1
q1 = Store['Sales_r'].quantile(0.25)
q3 = Store['Sales_r'].quantile(0.75)
iqr = q3 - q1


# Define the lower and upper bounds for the subset
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Remove the rows where the sales is outside the lower and upper bounds
Store = Store[(Store['Sales_r'] >= lower_bound) & (Store['Sales_r'] <= upper_bound)]

# Print summary statistics
print(Store['Sales_r'].describe())
```

```
# Calculate the interquartile range for the Profit_r column
Q1 = Store['Profit_r'].quantile(0.25)
Q3 = Store['Profit_r'].quantile(0.75)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to keep only the values within the lower and upper bounds
Store = Store[(Store['Profit_r'] >= lower_bound) & (Store['Profit_r'] <= upper_bound)]

# Print summary statistics
print(Store['Profit_r'].describe())
```

## Data Modelling

### Regression

Linear regression is a statistical modeling technique used to determine the relationship between one or more independent variables and a dependent variable. In other words, it is used to forecast the value of a dependent variable based on one or more independent variables. Linear regression assumes that the dependent variable and the independent variables have a linear relationship (Han, 2011).

Linear regression is a popular modeling technique in the retail industry because it allows businesses to identify the factors that contribute the most to their profitability. Retail stores can optimize their operations to increase revenue and decrease costs by understanding which variables have the greatest impact on profits. The retail store may use linear regression to determine which products or categories of products generate the most revenue, and then adjust their inventory and marketing strategies accordingly.

> MODEL 1

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#Load the data
X = Store[['Discount','Quantity','Sales_r']]
y = Store['Profit_r']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
# Create the linear regression object
model = LinearRegression()
# Fit the model to the training data
model.fit(X_train, y_train)
# Predict the target variable using the test data
y_pred = model.predict(X_test)

from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
x_train_scaled= scaler.transform(X_train)
x_test_scaled= scaler.transform(X_test)

# Print the coefficient and intercept of the model
print ("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)

# Evaluate the model' s performance using mean squared error
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('R^2 score:',r2)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print('Mean Squared Error:',mse)


#Create a scttered plot of Actual cs Predicted Values
plt.scatter(y_test,y_pred)

# Add a line showing the predicted values
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2,c='red')

# Set axis labels and plot title
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs Predicted Profit')

# Show the plot
plt.show()
```

> MODEL 2

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Load the data
X = Store[['Sales_r','State','Quantity','Sub-Category','Segment']]
y = Store['Profit_r']

#Convert X into dataframe

X=pd.DataFrame(X)

# Step 2 - Encode Categorical Data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers=[('State',OneHotEncoder(),['State']),
                                    ('SubCat',OneHotEncoder(),['Sub-Category']),
                                    ('Seg',OneHotEncoder(),['Segment'])]
                                   , remainder='passthrough')
X = ct.fit_transform(X)


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
# Create the linear regression object
model = LinearRegression()
# Fit the model to the training data
model.fit(X_train, y_train)
# Predict the target variable using the test data
y_pred = model.predict(X_test)

# Print the coefficient and intercept of the model
print ("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
# Evaluate the model' s performance using mean squared error
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('R^2 score:',r2)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print('Mean Squared Error:',mse)


#Create a scttered plot of Actual cs Predicted Values
plt.scatter(y_test,y_pred)

# Add a line showing the predicted values
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, c='red')

# Set axis labels and plot title
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs Predicted Profit')

# Show the plot
plt.show()
```

The above code provided is for predicting profit based on a store's sales, state, quantity, sub-category, and segment data. The code first loads the data and then uses OneHotEncoder to encode the categorical variables. The data is then divided into training and testing sets, and a linear regression object is created to fit the model to the training data. Using the test data, the model is then used to predict the target variable. The model's performance is assessed using the R2 score and mean squared error (Lipovetsky, 2021). Finally, a scatter plot is created to compare the target variable's actual and predicted values. The code can be used to analyze and predict the relationship between the input variables and the store's profit.

Comparing to the previous plot, the regression line is more close towards the actual values and has greater R^2 and lesser mse value. This suggests that the feature selection of the second model is a better fit for the given data.
